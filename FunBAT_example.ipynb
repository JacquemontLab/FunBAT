# Load the example datasets 

```python
pheno_df = pd.read_csv("data/example_pheno.tsv", sep="\t")
display(pheno_df.head())
```

```python
geno_df = pd.read_csv("data/example_geno.tsv", sep="\t")
display(geno_df.head())
```

```python
gene_sets_df = pd.read_csv("data/example_gene_sets_file.tsv", sep="\t")
display(gene_sets_df.head())

gene_dict = {}
for index, row in gene_sets_df.iterrows():
    gene_set_name = row['gene_set']
    genes = row['genes'].split(";")  # Assuming the genes are separated by semicolons
    gene_dict[gene_set_name] = genes

gene_list = gene_dict['cell_type_1']
```
